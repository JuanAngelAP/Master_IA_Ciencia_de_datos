```python
import pandas as pd
```

```python
#Cargar el archivo CSV
df = pd.read_csv("Titanic-Dataset.csv")

#Mostrar primeras información del df
df.info()
```

```python
#Motrar primeros y últimos datos
df
```

```python
#Total de datos faltantes
df.isnull().sum()
```

Manejo datos nulos de "Age"

```python
df['Age'].isnull().sum()
#Calcula mediana de edades
medianaAge = df['Age'].median()
print(medianaAge)
#Reemplaza valores faltantes por la mediana
df['Age'].fillna(medianaAge, inplace=True)

df.isnull().sum()
```

Manejo datos nulos de "Cabin"

```python
#Reemplazar los valores nulos por "Desconocido"
df['Cabin'].fillna("Desconocido", inplace=True)

df.isnull().sum()
```

Manejo datos nulos de "Embarked"

```python
#Calculo moda
modaEmbarked = df["Embarked"].mode()
print(modaEmbarked)
#Moda devuelve un objeto, por eso [0]
df['Embarked'].fillna(modaEmbarked[0], inplace=True)
df.isnull().sum()
```

Obtener datos a partir de la información del dataframe

```python
#Cantidad de supervivientes
supervivientes = df[df["Survived"] == 1]
print(len(supervivientes))
```

```python
#Mujeres que sobrevivieron
mujeres_sobrevivieron = df[(df["Survived"]==1) & (df["Sex"] == "female")]
print(len(mujeres_sobrevivieron))
```

```python
#Hombres que sobrevivieron
hombres_sobrevivieron = df[(df["Survived"]==1) & (df["Sex"] == "male")]
print(len(hombres_sobrevivieron))
```

```python
total = len(mujeres_sobrevivieron) + len(hombres_sobrevivieron)
print(f"Total de sobrevivientes: {total}")
```

```python
jovenes = df[(df["Age"] >= 0) & (df["Age"] <= 17)]
print(len(jovenes))
```

```python
adultos = df[(df["Age"] >= 18) & (df["Age"] <= 50)]
print(len(adultos))
```

```python
adultos_mayores = df[df["Age"] <= 50]
print(len(adultos_mayores))
```

```python
jovenes_supervivientes = jovenes[jovenes["Survived"] == 1]
print(len(jovenes_supervivientes))
```

```python
adultos_supervivientes = adultos[adultos["Survived"] == 1]
print(len(adultos_supervivientes))
```

Introducción a gráficas con MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
# contenido 
total_adultos = len(adultos)
total_adultos_supervivientes = len(adultos_supervivientes)
#Datos para la grafica
etiquetas =  ['Adultos', 'Adultos Supervivientes']
valores = [total_adultos,total_adultos_supervivientes]
#Creagar Grafica 
plt.figure(figsize=(6,4))
plt.bar(etiquetas,valores, color = ['gray','green'])
plt.title("Compracion: Adultos Totales vs Adultos Supervivientes")
plt.ylabel("Numero de personas")
plt.xlabel("Categorias")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# contenido 
total_adultos = len(adultos)
total_adultos_supervivientes = len(adultos_supervivientes)
total_no_supervivientes = total_adultos - total_adultos_supervivientes 
#Datos para la grafica
etiquetas =  ['Adultos Supervivientes', 'Adultos No Supervivientes']
valores = [total_adultos_supervivientes,total_no_supervivientes]
colores = ['#4CAF50','#E57373']
#Creagar Grafica 
plt.figure(figsize=(6,4))
plt.pie(
    valores,
    labels=etiquetas,
    colors=colores,
    shadow=True,
    autopct='%1.1f%%',
    startangle=90,
    explode=(0.05,0)
    )
plt.title("Compracion: Adultos No Supervivientes vs Adultos Supervivientes")
plt.show()
```
